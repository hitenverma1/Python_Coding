```python
#question 14

#what will be Entropy for the given data set?

# A = {x,x,x,y,y,y,y,y,y}

import numpy as np
from scipy.stats import entropy

# Define the dataset
dataset = ['x', 'x', 'x', 'y', 'y', 'y', 'y', 'y', 'y']

# Calculate the proportion of each value in the dataset
proportions = np.unique(dataset, return_counts=True)[1] / len(dataset)

# Calculate the entropy of the dataset using the proportions
entropy_value = entropy(proportions, base=2)

# Print the result
print('The entropy of the dataset is:', entropy_value)
```
