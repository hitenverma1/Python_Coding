```python
import pandas as pd
from faker import Faker
import random
from datetime import date, timedelta, datetime

# Initialize Faker to generate fake data
fake = Faker()
# Get today's date
today = date.today()
# Generate random data for the DataFrame
data = []
for _ in range(25):
    # customer_id = fake.uuid4()
    customer_id = fake.random_int(min=111111, max=999999)
    cust_name = fake.name()
    age = random.randint(18, 70)
    dob = today - timedelta(days=age*365)
    bank_ac_number = fake.random_int(min=1000000000, max=9999999999)
    bank_code = fake.random_int(min=1000, max=9999)
    balance_amount = round(random.uniform(100, 10000), 2)
    
    data.append([customer_id, cust_name, age, dob, bank_ac_number, bank_code, balance_amount])

# Create the DataFrame
columns = ['customer_id', 'cust_name', 'age', 'dob', 'bank_ac_number', 'bank_code', 'balance_amount']
df1 = pd.DataFrame(data, columns=columns)

df1
```

```python
df = df1.copy()
```
