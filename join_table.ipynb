```python
import pandas as pd
import numpy as np
import datetime
```

```python
students_info = pd.DataFrame({"student_id" : [101,221,782,902,224,972,322,136,131,424],
            "student_name":["sagar","aman","rahul","hiten","sumit","varun","hardeep","manav","shubham","akshay"]})
```

```python
marks_info = pd.DataFrame(columns=("student_id","subject_name","marks_scored","total_marks","date_of_exam"),
             data= ((221,"maths",67,100,"2022-09-12"),
                    (902,"hindi",56,100,"2022-04-11"),
                    (221,"economics",54,100,"2022-07-15"),
                    (131,"accounts",87,100,"2022-05-16"),
                    (221,"maths",67,100,"2022-03-23"),
                    (972,"english",75,100,"2022-01-22"),
                    (424,"accounts",87,100,"2022-02-26"),
                    (782,"science",86,100,"2022-07-04"),
                    (224,"economics",65,100,"2022-01-03"),
                    (902,"economics",91,100,"2022-02-29"),
                    (221,"maths",75,100,"2022-06-11"),
                    (902,"economics",73,100,"2022-03-10"),
                    (136,"hindi",83,100,"2022-09-12"),
                    (101,"maths",81,100,"2022-05-04"),
                    (972,"english",59,100,"2022-03-01")))
```

```python
print(students_info.shape)
print(marks_info.shape)
```

```python
merge = pd.merge(students_info,marks_info, on="student_id",how="inner")
```

```python
merge.info()
```

```python
#fetch the data of last exam given by each students
df1 = merge.sort_values(by="date_of_exam", ascending=False)
df1.drop_duplicates(subset='student_name')
```

```python
pd.merge(students_info,marks_info, on="student_id",how="inner")
```

```python
# Calculate the percentage for each student
df23 = merge.groupby('student_name').sum('marks_scored')
df23['percentage'] = round((df23['marks_scored']/df23['total_marks'])*100,2)
```

```python
df23
```

```python
# Group the merged dataframe by "id" and "student_name"
grouped_df = merge.groupby(["student_id", "student_name"])
# Calculate the percentage for each student
result_df = grouped_df.apply(lambda x: (x["marks_scored"].sum() / x["total_marks"].sum()) * 100)
result_df = result_df.reset_index()
result_df.columns = ["student_id", "student_name", "percentage"]
```
