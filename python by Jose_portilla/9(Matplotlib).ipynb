```python
a = np.linspace(0,10,11)
b = a**4
```

```python
x = np.arange(0,10)
y = 2*x
```

```python
fig,axes = plt.subplots (nrows = 1, ncols = 2)
axes [0].plot(x,y)
axes [1].plot(a,b)
plt.tight_layout()
```

```python
fig,axes = plt.subplots (nrows = 3, ncols = 1)
axes [0].plot(x,y)
plt.tight_layout()
```

```python
fig,axes = plt.subplots(nrows = 2,ncols = 2)
axes[0][0].plot(x,y)
axes[0][1].plot(x,y)
axes[1][0].plot(x,y)
axes[1][0].set_ylabel('Y LABEL 1,0')
axes[1][0].set_xlim(2,6)
axes[1][1].plot(a,b)
axes[1][1].set_title('Title1,1')
fig.suptitle("Figure Level",fontsize = 16)
plt.tight_layout()
fig.set_figwidth(10)
fig.savefig ('new_subplot.png',bbox_inches = 'tight')
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x,label='x vs x')
ax.plot(x,x**2,label='x vs x^2')
ax.legend(loc=(1.1,0.5))
```

```python
#colors & style
```

```python
x = np.linspace(0,11,10)
fig  = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x,color = 'purple',marker='s')
ax.plot(x,x+2,color='#ffc800',ls='--',lw=12)
```

```python
x = np.linspace(0,11,10)
fig  = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x+2,color='#ffc800',marker = 'o', ls='--',lw=2,
       markerfacecolor='red',markeredgewidth=20,markeredgecolor='orange')
```
