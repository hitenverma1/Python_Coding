```python
df= pd.read_csv("Sales_Funnel_CRM.csv")
```

```python
df
```

```python
lic = df[['Company','Product','Licenses']]
```

```python
lic
```

```python
df.pivot(index='Company',columns='Product',values='Licenses')
```

```python
pd.pivot(data = lic,index='Company',columns='Product',values='Licenses')
```

```python
pd.pivot_table(df,index="Company", aggfunc='sum')
```

```python
pd.pivot_table(df,index="Company", aggfunc='sum',values=['Licenses','Sale Price'])
```

```python
df.groupby('Company').sum()
```

```python
pd.pivot_table(df,index=['Account Manager','Contact'], aggfunc='sum',values='Sale Price')
```

```python
pd.pivot_table(df,index=['Account Manager','Contact'], aggfunc='sum',values='Sale Price',columns = ['Product'])
```

```python
pd.pivot_table(df,index=['Account Manager','Contact','Product'], aggfunc=[np.sum],values='Sale Price')
```
